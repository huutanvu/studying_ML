## How to sample from a Gaussian distribution

```python
r = 10*np.random.randn(10000) + 5 # Mean = 5
```

```python
plt.hist(r, bins=100);
```

```python
r = np.random.randn(10000, 2)
```

```python
plt.scatter(r[:, 0], r[:, 1])
```

```python
r[:, 1] = 2*r[:, 1] + 2 # 5 as standard deviation + 2 mean
```

```python
plt.scatter(r[:, 0], r[:, 1])
plt.axis('equal')
```

```python
cov = np.array([[1, 0.8], [0.8, 3]])
```

```python
from scipy.stats import multivariate_normal as mvn
```

```python
mu = np.array([0, 2])
```

```python
r = mvn.rvs(mean=mu, cov=cov, size=1000)
```

```python
r
```

```python
plt.scatter(r[:, 0], r[:, 1])
```

```python
r = np.random.multivariate_normal(mean=mu, cov=cov, size=1000)
```

```python
plt.scatter(r[:, 0], r[:, 1])
```
