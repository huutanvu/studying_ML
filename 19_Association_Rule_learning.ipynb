### How the Apriori algorithm works?
- The Apriori algorithm has 3 parts:
#### Support:
<img src="Images/15_1_support.JPG" height="300">
- Like: how many people will watch movie M out of those people. It means, how many people support (or like) movie M.
#### Confidence:
- For movie: the number of people who have seen movie M1 also watch movie M2 in the whole set of people who watch movie M1.
- For market: the number of people who bought item I1 also bought I2 in the whole set of people who bought item I1.
<img src="Images/15_1_confidence.JPG" height="300">
#### Lift:
<img src="Images/15_1_lift.JPG" height="300">

### Algorithm:
<img src="Images/15_1_algorithm.JPG" height="300">
- This is a slow algorithm because it has to go through all the pairs to compare. That's why we have to set a minimum support and confidence.
- Strongest rule with highest lift will be looked into first.
